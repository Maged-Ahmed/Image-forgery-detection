```python
import seaborn as sns
import tensorflow as tf
import cv2
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (256,256,3)),
    tf.keras.layers.Dense(250, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1, activation='softmax'),
])

model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.SGD(),
    metrics = [tf.keras.metrics.Accuracy()]
)
```

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(validation_split=0.2) # set validation split

train_generator = train_datagen.flow_from_directory(
    '/kaggle/input/casia-dataset/CASIA1',
    subset='training',
target_size=(256,256),
class_mode="binary") #training data
    
test_generator = train_datagen.flow_from_directory(
    '/kaggle/input/casia-dataset/CASIA1', # same directory as training data
    subset='validation',
target_size=(256,256),
class_mode="binary") #validation data
```

```python
train = model.fit(train_generator,epochs=20,batch_size=300,validation_data=test_generator, validation_steps=50)
```
